```python
xRange = [-10,10]
f = lambda x: x**2 + x**3 + x
x = np.linspace(xRange[0], xRange[1], num = 25) # = for i in range(points): f(points)

plt.title('test\nSubtitle')
plt.xlabel('x')
plt.ylabel('f(x)')


plt.plot(x, f(x),
         'o', #empty to continuos  functio, 'o' to draw points
         label = 'curquita',
         color = 'red')
plt.legend()
plt.grid()

plt.show()#Is not allways necesary
```

```python
minX = -10
maxX = 10
f = lambda x:(x**2)
g = lambda x:(1/x)

axes = plt.gca() #axes limits
axes.set_xlim([minX,maxX])
axes.set_ylim([minX,maxX])

x1 = np.linspace(minX, maxX, num = 1000)
x2 = np.linspace(minX, maxX, num = 1000)
plt.plot(x1 ,f(x1), color = 'r')
plt.plot(x2, g(x2), color = 'purple')
plt.grid()
plt.show()
```

```python
def limitAxes(x1, x2, y1 = 0, y2 = 0):
    if(y1 == 0 and y2 == 0):
        y1 = x1; y2 = x2
    axes = plt.gca()
    axes.set_xlim([x1,x2])
    axes.set_ylim([y1,y2])
    
limitAxes(-20,40,-100,1)
plt.grid()
plt.show()
limitAxes(-20,799)
plt.grid()
plt.show()
```
