```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
#Did in .py there exist more options while doing a plt
```

```python
x = [1,4,8,16] 
y = [1,2,3,4]
plt.plot(x,y)
```

```python
xRange = [-10,10]
f = lambda x: x**2 + x**3 + x
x = np.linspace(xRange[0], xRange[1], num = 25) # = for i in range(points): f(points)

plt.title('test\nSubtitle')
plt.xlabel('x')
plt.ylabel('f(x)')


plt.plot(x, f(x),
         'o', #empty to continuos  functio, 'o' to draw points
         label = 'curquita',
         color = 'red')
plt.legend()
plt.grid()

plt.show()#Is not allways necesary
```

```python
minX = -10
maxX = 10
f = lambda x:(x**2)
g = lambda x:(1/x)

axes = plt.gca() #axes limits
axes.set_xlim([minX,maxX])
axes.set_ylim([minX,maxX])

x1 = np.linspace(minX, maxX, num = 1000)
x2 = np.linspace(minX, maxX, num = 1000)
plt.plot(x1 ,f(x1), color = 'r')
plt.plot(x2, g(x2), color = 'purple')
plt.grid()
plt.show()
```

```python
def limitAxes(x1, x2, y1 = 0, y2 = 0):
    if(y1 == 0 and y2 == 0):
        y1 = x1; y2 = x2
    axes = plt.gca()
    axes.set_xlim([x1,x2])
    axes.set_ylim([y1,y2])
    
limitAxes(-20,40,-100,1)
plt.grid()
plt.show()
limitAxes(-20,799)
plt.grid()
plt.show()
```

```python
minX, maxX = -10,5
bars = 15

f = lambda x: x**2
x = np.linspace(minX,maxX, num = bars)

plt.bar(x, f(x),
        label = 'x^2 in bars')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
import random as r

val1 = []
val2 = []
bin = []; binr = 13
vlen = 30
ran = 100

for i in range(0, vlen):
    val1.append(r.randint(0,ran))
    val2.append(i)
    if(i<binr):
        bin.append(i*10)

#plt.bar(val2,val1)#meh
plt.hist(val1, bin,
         histtype = 'bar',
         rwidth = 0.9)

plt.show()
```

```python
xRange = [-10,10]
f = lambda x: x**2 + x**3 + x
x = np.linspace(xRange[0], xRange[1])

plt.scatter(x, f(x), marker = 'x', s = 300) # to search for correlations
plt.show()
```

```python
minX, maxX = 0,5
x = np.linspace(minX, maxX, num = maxX - minX)

print('independent = x, red = v1, green = v2, blue = v3') 
#or
plt.plot([],[],label = 'v1', color = 'r', linewidth = 10)
plt.plot([],[],label = 'v2', color = 'g', linewidth = 10)
plt.plot([],[],label = 'v3', color = 'b', linewidth = 10)
plt.stackplot(x, [10,7,6,10,9], [0,2,3,1,3], [5,4,6,6,6],
              colors = ['r', 'g', 'b'])
plt.legend()
plt.show()
```
